```python
import pandas as pd

# Load the dataset
file_path = 'Data_GWPZ_Arif_without_coordinate.xls'
data = pd.read_excel(file_path)

# Display the first few rows of the dataset to understand its structure
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# Creating a copy of the dataset to apply transformations
data_encoded = data.copy()

# Initializing label encoder
label_encoder = LabelEncoder()

# List of categorical columns to be encoded
categorical_columns = ['GWPZ_Class', 'Drainage_Density_Class', 'Geology_Class',
                       'Geomorphology_Class', 'Lineament_Density_Class', 'LULC_Class',
                       'NDVI_Class', 'Slope_Class', 'TWI_Class', 'Soil_Texture_Class']

# Applying label encoding to each categorical column
for column in categorical_columns:
    data_encoded[column] = label_encoder.fit_transform(data_encoded[column])

# Display the first few rows of the transformed dataset
data_encoded.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm, skew

# Check for missing values
missing_values = data_encoded.isnull().sum()

# Data types of each column
data_types = data_encoded.dtypes

# Normality check
# For this, let's plot histograms and calculate skewness for numerical columns
numerical_columns = data_encoded.select_dtypes(include=['int64', 'float64']).columns

# Plotting histograms and calculating skewness
skewness = {}
for column in numerical_columns:
    skewness[column] = skew(data_encoded[column])
    sns.distplot(data_encoded[column], fit=norm)
    plt.title(f'Normality Check for {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()

# Normalizing the dataset
# Using Min-Max Scaling as an example
from sklearn.preprocessing import MinMaxScaler

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Scale the dataset (excluding missing values handling for now)
data_normalized = pd.DataFrame(scaler.fit_transform(data_encoded), columns=data_encoded.columns)

# Check the first few rows of normalized data
normalized_head = data_normalized.head()

missing_values, data_types, skewness, normalized_head
```

```python
# Calculating descriptive statistics for the dataset
descriptive_statistics = data_encoded.describe()

descriptive_statistics
```

```python
# Plotting histograms for each variable in the dataset
for column in data_encoded.columns:
    plt.figure(figsize=(10, 4))
    sns.histplot(data_encoded[column], kde=False, bins=20)
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Plotting boxplots for each variable in the dataset
for column in data_encoded.columns:
    plt.figure(figsize=(10, 4))
    sns.boxplot(x=data_encoded[column])
    plt.title(f'Boxplot of {column}')
    plt.xlabel(column)
    plt.show()
```

```python
# Creating a heatmap to visualize the correlation matrix of the dataset
plt.figure(figsize=(12, 8))
correlation_matrix = data_encoded.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Heatmap of Correlation Matrix')
plt.show()
```

```python
import numpy as np
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Correlation Coefficients
correlation_matrix = data_encoded.corr()

# Variance Inflation Factor (VIF) for Multicollinearity
# The VIF estimates how much the variance of a regression coefficient is inflated due to multicollinearity.
# VIF starts at 1 and has no upper limit. A value of 1 indicates no correlation between a given variable and any others.
# VIF values above 10 may indicate high multicollinearity.

# Preparing data for VIF calculation
X = data_encoded.assign(const=1)  # Adding a constant column for VIF calculation

# Calculating VIF for each variable
vif_data = pd.DataFrame()
vif_data["Variable"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

correlation_matrix, vif_data
```

```python
# Performing Group-by analysis on the dataset

# Example: Grouping by 'GWPZ_Class' and calculating mean values for other variables
grouped_data_mean = data_encoded.groupby('GWPZ_Class').mean()

# Grouping by 'GWPZ_Class' and calculating sum values for other variables
grouped_data_sum = data_encoded.groupby('GWPZ_Class').sum()

# Grouping by 'GWPZ_Class' and calculating count for each group
grouped_data_count = data_encoded.groupby('GWPZ_Class').size()

grouped_data_mean, grouped_data_sum, grouped_data_count
```

```python
# Preparing the grouped data for output
grouped_data_output = pd.ExcelWriter('grouped_data_analysis.xlsx')

# Writing the grouped data to the Excel file
grouped_data_mean.to_excel(grouped_data_output, sheet_name='Grouped_Mean')
grouped_data_sum.to_excel(grouped_data_output, sheet_name='Grouped_Sum')
grouped_data_count.to_frame(name='Count').to_excel(grouped_data_output, sheet_name='Grouped_Count')

# Saving the Excel file
grouped_data_output.save()

# Providing the file path for download
file_path = 'grouped_data_analysis.xlsx'
file_path
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Selecting numerical columns (excluding 'GWPZ_Class' which is categorical)
numerical_data = data_encoded.drop('GWPZ_Class', axis=1)

# Initializing PCA
pca = PCA()

# Fitting and transforming the data
pca_data = pca.fit_transform(numerical_data)

# Variance explained by each principal component
explained_variance = pca.explained_variance_ratio_

# Cumulative variance explained
cumulative_variance = np.cumsum(explained_variance)

# Plotting the explained variances
plt.figure(figsize=(10, 6))
plt.bar(range(1, len(explained_variance) + 1), explained_variance, alpha=0.5, align='center', label='Individual explained variance')
plt.step(range(1, len(cumulative_variance) + 1), cumulative_variance, where='mid', label='Cumulative explained variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal component index')
plt.legend(loc='best')
plt.title('Explained Variance by Different Principal Components')
plt.show()

explained_variance, cumulative_variance
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Defining the independent variables (X) and the dependent variable (Y)
X = data_encoded.drop('GWPZ_Class', axis=1)
Y = data_encoded['GWPZ_Class']

# Splitting the dataset into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)

# Creating a Linear Regression model
lr_model = LinearRegression()

# Fitting the model to the training data
lr_model.fit(X_train, Y_train)

# Making predictions on the test set
Y_pred = lr_model.predict(X_test)

# Calculating the Mean Squared Error and R-squared value
mse = mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)

# Coefficients and intercept
coefficients = lr_model.coef_
intercept = lr_model.intercept_

mse, r2, coefficients, intercept
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Since Logistic Regression requires binary classification, let's modify 'GWPZ_Class' to binary
# For simplicity, we will classify 'High' and 'Very High' as 1, and the rest as 0.
# Adjust this according to your specific classification needs.
binary_Y = Y.apply(lambda x: 1 if x in [3, 4] else 0)  # Assuming classes 3 and 4 are 'High' and 'Very High'

# Splitting the dataset into training and testing sets for the binary classification
X_train, X_test, binary_Y_train, binary_Y_test = train_test_split(X, binary_Y, test_size=0.3, random_state=42)

# Creating a Logistic Regression model
logistic_model = LogisticRegression(max_iter=1000)

# Fitting the model to the training data
logistic_model.fit(X_train, binary_Y_train)

# Making predictions on the test set
binary_Y_pred = logistic_model.predict(X_test)

# Calculating accuracy, confusion matrix, and classification report
accuracy = accuracy_score(binary_Y_test, binary_Y_pred)
conf_matrix = confusion_matrix(binary_Y_test, binary_Y_pred)
class_report = classification_report(binary_Y_test, binary_Y_pred)

accuracy, conf_matrix, class_report
```

```python
from sklearn.tree import DecisionTreeClassifier

# Creating a Decision Tree model
dt_model = DecisionTreeClassifier(random_state=42)

# Fitting the model to the training data
dt_model.fit(X_train, binary_Y_train)

# Making predictions on the test set
binary_Y_pred_dt = dt_model.predict(X_test)

# Calculating accuracy, confusion matrix, and classification report
accuracy_dt = accuracy_score(binary_Y_test, binary_Y_pred_dt)
conf_matrix_dt = confusion_matrix(binary_Y_test, binary_Y_pred_dt)
class_report_dt = classification_report(binary_Y_test, binary_Y_pred_dt)

accuracy_dt, conf_matrix_dt, class_report_dt
```

```python
from sklearn.ensemble import RandomForestClassifier

# Creating a Random Forest model
rf_model = RandomForestClassifier(random_state=42)

# Fitting the model to the training data
rf_model.fit(X_train, binary_Y_train)

# Making predictions on the test set
binary_Y_pred_rf = rf_model.predict(X_test)

# Calculating accuracy, confusion matrix, and classification report
accuracy_rf = accuracy_score(binary_Y_test, binary_Y_pred_rf)
conf_matrix_rf = confusion_matrix(binary_Y_test, binary_Y_pred_rf)
class_report_rf = classification_report(binary_Y_test, binary_Y_pred_rf)

accuracy_rf, conf_matrix_rf, class_report_rf
```

```python
from sklearn.ensemble import GradientBoostingClassifier

# Creating a Gradient Boosting Classifier model
gbm_model = GradientBoostingClassifier(random_state=42)

# Fitting the model to the training data
gbm_model.fit(X_train, binary_Y_train)

# Making predictions on the test set
binary_Y_pred_gbm = gbm_model.predict(X_test)

# Calculating accuracy, confusion matrix, and classification report
accuracy_gbm = accuracy_score(binary_Y_test, binary_Y_pred_gbm)
conf_matrix_gbm = confusion_matrix(binary_Y_test, binary_Y_pred_gbm)
class_report_gbm = classification_report(binary_Y_test, binary_Y_pred_gbm)

accuracy_gbm, conf_matrix_gbm, class_report_gbm
```

```python
from sklearn.svm import SVC

# Creating a Support Vector Machine (SVM) model
svm_model = SVC(random_state=42)

# Fitting the model to the training data
svm_model.fit(X_train, binary_Y_train)

# Making predictions on the test set
binary_Y_pred_svm = svm_model.predict(X_test)

# Calculating accuracy, confusion matrix, and classification report
accuracy_svm = accuracy_score(binary_Y_test, binary_Y_pred_svm)
conf_matrix_svm = confusion_matrix(binary_Y_test, binary_Y_pred_svm)
class_report_svm = classification_report(binary_Y_test, binary_Y_pred_svm)

accuracy_svm, conf_matrix_svm, class_report_svm
```

```python
from sklearn.neural_network import MLPClassifier

# Creating a Neural Network (Multi-Layer Perceptron) model
nn_model = MLPClassifier(hidden_layer_sizes=(100,), max_iter=1000, random_state=42)

# Fitting the model to the training data
nn_model.fit(X_train, binary_Y_train)

# Making predictions on the test set
binary_Y_pred_nn = nn_model.predict(X_test)

# Calculating accuracy, confusion matrix, and classification report
accuracy_nn = accuracy_score(binary_Y_test, binary_Y_pred_nn)
conf_matrix_nn = confusion_matrix(binary_Y_test, binary_Y_pred_nn)
class_report_nn = classification_report(binary_Y_test, binary_Y_pred_nn)

accuracy_nn, conf_matrix_nn, class_report_nn
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay

# List of model names and their corresponding confusion matrices
model_names = ['Logistic Regression', 'Decision Tree', 'Random Forest', 'Gradient Boosting', 'SVM', 'Neural Network']
confusion_matrices = [conf_matrix, conf_matrix_dt, conf_matrix_rf, conf_matrix_gbm, conf_matrix_svm, conf_matrix_nn]

# Plotting confusion matrices
plt.figure(figsize=(15, 10))

for i, (model_name, conf_matrix) in enumerate(zip(model_names, confusion_matrices)):
    plt.subplot(2, 3, i+1)
    ConfusionMatrixDisplay(conf_matrix).plot(cmap='Blues', ax=plt.gca())
    plt.title(model_name)

plt.tight_layout()
plt.show()
```
